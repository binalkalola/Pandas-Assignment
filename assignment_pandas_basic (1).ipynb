Question 7. select the rows with columns ['birds', 'visits'] where the age is

Answer : birds['age'].isna() this checks where age is missing(NaN)

question 8. Select the rows where the birds is a Cranes and the age is less
            than 4
Answer :(birds['birds'] == 'Cranes') checks each row if birds is cranes is true
        otherwise false
        (birds['age] < 4) if age is less than 4 is true otherwise false
        &(AND operater) combine both condition
        only row where both are true will be selected
        birds[ ... ] filters the Dataframe based on the combined condition


```python
print(birds[(birds['birds'] == 'Cranes') & (birds['age'] < 4)])

```

Question 14. Sort dataframe (birds) first by the values in the 'age' in
              decending order, then by the value in the 'visits' column in ascending order.
Answer : sort_values() function is used to sort the DataFrame
          by=['age', 'visits'] first sort by age then sort by visits
          ascending=[False, True]

Question 15. Replace the priority column values with'yes' should be 1 and 'no'
            should be 0
  
Answer :birds['priority'] selects the priority column
        .map({...}) replaces values using a dictionary
        'yes' is replaced with 1
